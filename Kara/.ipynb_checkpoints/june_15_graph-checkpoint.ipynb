```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
june_15_csv = 'june_15.csv'
```

```python
june_15_df = pd.read_csv(june_15_csv)
```

```python
june_15_df.head()
```

```python
for x in june_15_df['Date']:
    june_15_df['Date'][x]
```

```python
date = june_15_df['Date'][0] 
date.day
```

```python
converted_timestamps = []
for raw in june_15_df['Date']:
    # https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
    # http://strftime.org/
    converted_time = datetime.strptime(raw, "%a %b %d %H:%M:%S %z %Y")
    converted_timestamps.append(converted_time)
```

```python
converted_timestamps[0].day
```

```python
day = []
for x in range(len(converted_timestamps)):
    days = converted_timestamps[x].day
    day.append(days)
```

```python
june_15_df['Day'] = day
```

```python
june_15_df.head()
```

```python
only_june_15_df = june_15_df.loc[june_15_df['Day'] == 15, :]
```

```python
only_june_15_df.head()
```

```python
len(only_june_15_df['Day'])
```

```python
mean_per_hour = june_15_df.groupby('Hour').mean()
mean_per_hour
```

```python
mean_per_hour.reset_index()
```

```python
plt.scatter(range(len(mean_per_hour['Compounded'] )), mean_per_hour['Compounded'])
```
