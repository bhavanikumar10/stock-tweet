```python
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
```

```python
fb_18 = 'june_18_FB.csv'
nin_18_1 = 'june_18_N_1.csv'
nin_18_2 = 'june_18_N_2.csv'
mb_18 = 'june_18_MB_1.csv'
nin_19 = 'june_19_N_1.csv'
```

```python
fb_18_df = pd.read_csv(fb_18)
nin_18_1df = pd.read_csv(nin_18_1)
nin_18_2df = pd.read_csv(nin_18_2)
mb_18_df = pd.read_csv(mb_18)
nin_19_df = pd.read_csv(nin_19)
```

```python
fb_18_df.head()
```

```python
converted_timestamps = []
for raw in fb_18_df['Date']:
    # https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
    # http://strftime.org/
    converted_time = datetime.strptime(raw, "%a %b %d %H:%M:%S %z %Y")
    converted_timestamps.append(converted_time)
```

```python
day = []
for x in range(len(converted_timestamps)):
    days = converted_timestamps[x].day
    day.append(days)
```

```python
fb_18_df['Day'] = day
fb_18_df.head()
```

```python
fb_18_df = fb_18_df.loc[fb_18_df['Day'] == 18, :]
nin_18_1df = nin_18_1df.loc[nin_18_1df['Day'] == 18, :]
nin_18_2df = nin_18_2df.loc[nin_18_2df['Day'] == 18, :]
nin_19_df = nin_19_df.loc[nin_19_df['Day'] == 19, :]
mb_18_df = mb_18_df.loc[mb_18_df['Day'] == 18, :]
```

```python
nin_18_1df.head()
```

```python
fb_18_mean = fb_18_df.groupby('Hour').mean()
fb_18_mean
```

```python
nin_18_mean1 = nin_18_1df.groupby('Hour').mean()
nin_18_mean2 = nin_18_2df.groupby('Hour').mean()
nin_19_mean = nin_19_df.groupby('Hour').mean()
mb_18_mean = mb_18_df.groupby('Hour').mean()
```

```python
nin_18_mean1
```

```python
plt.plot(range(len(fb_18_mean['Compounded'][13:21] )), fb_18_mean['Compounded'][13:21], "-o", label='Compounded')

plt.xticks(range(len(fb_18_mean['Compounded'][13:21] )), [9,10,11,12,1,2,3,4])
plt.xlabel('Time(EDT)')
plt.ylabel('Compound Score')
plt.title('Sentiment Analysis of Facebook Mentions (June 18th)')
plt.savefig('FB_18.png', bbox_inches='tight')
```
